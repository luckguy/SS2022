### 複數弦波訊號

複數弦波訊號 $x(t)$ 係由 [複數指數函數](https://en.wikipedia.org/wiki/Exponential_function#Complex_plane) 所定義

\begin{equation}
x(t) = e^{s t} 
\end{equation}

其中 $s = \alpha + j \omega_0$. $\alpha, \omega_0 \in \mathbb{R}$ 與 $j$ 為虛數符號 $(j^2=-1)$. 利用尤拉公式[Euler's formula](https://en.wikipedia.org/wiki/Euler's_formula) 可將上述 $x(t)$ 訊號改寫成

\begin{equation}
x(t) = e^{(\alpha + j \omega_0) t} = e^{\alpha t} \cos(\omega_0 t) + j e^{\alpha t} \sin(\omega_0 t)
\end{equation}

$\omega_0 = 2 \pi f_0$. 對於 $t>0$,    $e^{\alpha t}$ 決定訊號強弱

* $\alpha < 0$, 訊號指數遞減
* $\alpha = 0$, 訊號強弱不變
* $\alpha > 0$, 訊號指數遞增

至於 $t<0$ 則得到與 $t>0$ 相反之結果

### 一般化情況 

底下程式使用 Python 模組 [SymPy](http://docs.sympy.org/latest/index.html) 定義 複指數訊號 $x(t)=e^{st}=e^{j(\alpha+j\omega_0)t}$。變數 $t$, $\omega_0$ and $\alpha$ 均為實數變數。

```python
import sympy as sym
sym.init_printing()

t, alpha, omega0 = sym.symbols('t alpha omega', real=True)
s = alpha + 1j*omega0
x = sym.exp(s*t)
x
```

### Case 1: $s$ 為實數

```python
# 設定 alpha 與 sigma 數值
y = x.subs({omega0: 0, alpha: -.5})  # alpha=-0.5 

sym.plot(sym.re(y), (t, -1, 2*sym.pi), ylabel=r'$e^{st}$')
y = x.subs({omega0:0, alpha:0.5}) #alpha=0.5
p=sym.plot(sym.re(y), (t, -2*sym.pi,1), ylabel=r'Im{$e^{st}$}');
```

### Case 2: $s$ 為純虛數。$s=j\omega_0$
\begin{equation}
x(t)=e^{st}=e^{j\omega_0 t}=\cos(\omega_0 t)+j\sin(\omega_0 t)
\end{equation}

```python
# 設定 alpha 與 sigma 數值
y = x.subs({omega0: 10, alpha: 0.})  # alpha=0, omega=10

sym.plot(sym.re(y), (t, 0, 2*sym.pi), ylabel=r'Re{$e^{st}$}')
sym.plot(sym.im(y), (t, 0, 2*sym.pi), ylabel=r'Im{$e^{st}$}');
```
