### 複數弦波訊號

複數弦波訊號 $x(t)$ 係由 [複數指數函數](https://en.wikipedia.org/wiki/Exponential_function#Complex_plane) 所定義

\begin{equation}
x(t) = e^{s t} 
\end{equation}

其中 $s = \alpha + j \omega_0$. $\alpha, \omega_0 \in \mathbb{R}$ 與 $j$ 為虛數符號 $(j^2=-1)$. 利用尤拉公式[Euler's formula](https://en.wikipedia.org/wiki/Euler's_formula) 可將上述 $x(t)$ 訊號改寫成

\begin{equation}
x(t) = e^{(\alpha + j \omega_0) t} = e^{\alpha t} \cos(\omega_0 t) + j e^{\alpha t} \sin(\omega_0 t)
\end{equation}

The real/imaginary part of the exponential signal is given by a weighted cosine/sine with angular frequency $\omega_0 = 2 \pi f_0$. 對於 $t>0$,    $e^{\alpha t}$ 決定訊號強弱

* $\alpha < 0$, 訊號指數遞減
* $\alpha = 0$, 訊號強弱不變
* $\alpha > 0$, 訊號指數遞增

至於 $t<0$ 則得到與 $t>0$ 相反之結果

### 一般化情況 

底下程式使用 Python 模組 [SymPy](http://docs.sympy.org/latest/index.html) 定義 複指數訊號。變數 $t$, $\omega_0$ and $\alpha$ 均為變數.

```python
import sympy as sym
sym.init_printing()

t, alpha, omega0 = sym.symbols('t alpha omega', real=True)
s = alpha + j*omega0
x = sym.exp(s*t)
x
```

複指數訊號的指數部分 $s = \alpha + j \omega_0$。 其中  $\alpha$ 與 $\omega_0$ 為特定值.
