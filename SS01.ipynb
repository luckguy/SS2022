## 基本運算


### 定義訊號

```python
import sympy as sym
sym.init_printing()
t=sym.symbols('t',real=True)
x=(t-1)*(sym.Heaviside(t-2)-sym.Heaviside(t-1))+sym.Heaviside(t)-sym.Heaviside(t-2)
sym.plot(x,(t,-3,3),ylim=[-0.2,1.3],ylabel='x(t)',line_color='red')
```

### 時間位移



$x(t)$ 的位移訊號定義為

\begin{equation}
y(t) = x(t-\tau)
\end{equation}
with $\tau \in \mathbb{R}$. 

* $\tau > 0$, 往右移
* $\tau < 0$, 往左移

```python
x=(t-1)*(sym.Heaviside(t-2)-sym.Heaviside(t-1))+sym.Heaviside(t)-sym.Heaviside(t-2)
sym.plot(x,(t,-2,4),ylim=[-0.2,1.3],ylabel='x(t)',line_color='red')
```

### 位移範例 


\begin{equation}
x(t) = \text{rect}\left(t - \frac{1}{2} \right) + \frac{2}{3} \cdot \text{rect}\left(t-\frac{3}{2} \right) + \frac{1}{3} \cdot \text{rect} \left(t-\frac{5}{2} \right)
\end{equation}

```python
y=x.subs(t,t-1)
sym.plot(y,(t,-2,4),ylim=[-0.2,1.3],ylabel='x(t-1)',line_color='red')
```

### 時間刻度

The temporal scaling of a signal $x(t)$ is defined as

\begin{equation}
y(t) = x(a \cdot t)
\end{equation}

with $a \in \mathbb{R}$. The signal $x(t)$ is

* $0 < a < 1$ 延伸
* $a > 1$ 擠壓
* time-reversed and scaled for $a < 0$



**時間刻度範例**

The following example illustrates the temporal scaling of the staircase signal $y(t) = x(a \cdot t)$ introduced in the previous example. The original $x(t)$ is plotted in gray, the scaled signal $y(t)$ in blue. Here stretching is realized, such that $y(t)$ is twice as long as $x(t)$.

```python
a = 1/2
y = x.subs(t, a*t)


px = sym.plot(x, (t, -3, 7), ylim=[-0.2, 1.2],
              ylabel=r'$x(t)$', show=False, line_color='gray')
py = sym.plot(y, (t, -3, 7),
              ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False,line_color='red')
py.extend(px)
py.show()
```

**Example**
書本範例
\begin{equation}
   y(t)=x\left(-t+1\right)
\end{equation}


```python
tau = 1
y = x.subs(t, - t+tau)

px = sym.plot(x, (t, -2, 4), ylim=[-0.2, 1.2],
              ylabel=r'$x(t)$', show=False, line_color='gray')
py = sym.plot(y, (t, -2, 4), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False,line_color='red')
py.extend(px)
py.show()
```
